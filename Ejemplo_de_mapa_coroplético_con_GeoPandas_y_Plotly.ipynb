```python
import pandas as pd
import geopandas as gpd
import plotly.express as px
import pyproj
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
municipios = '/content/drive/MyDrive/Pruebas/Municipios_IGN.shp'
gpd_municipios = gpd.read_file(municipios)
```

```python
gpd_municipios
```

```python
gpd_alicante = gpd_municipios[gpd_municipios['CODNUT3'].str.contains('ES521', regex=True)]
```

```python
gpd_alicante.head()
```

```python
fig = px.choropleth(gpd_alicante,
                    geojson=gpd_alicante.geometry,
                    locations=gpd_alicante.index,
                    hover_name='NAMEUNIT'
                    )
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.show()
```

```python
habitantes = '/content/drive/MyDrive/Pruebas/Revision_Padron_Habitantes.xlsx'
df_habitantes = pd.read_excel(habitantes)
```

```python
df_habitantes.head()
```

```python
df_habitantes.columns
```

```python
df_habitantes.columns = ['INE','Municipio','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','Vacío','Hombres 2021','Mujeres 2021']
```

```python
df_habitantes = df_habitantes[['INE','Municipio','2021','Hombres 2021','Mujeres 2021']]
```

```python
df_habitantes = df_habitantes[df_habitantes['2021'].isna() == False]
```

```python
df_habitantes
```

```python
df_habitantes['INE'] = df_habitantes['INE'].astype('string')
```

```python
df_habitantes.dtypes
```

```python
df_habitantes['INE'] = '0' + df_habitantes['INE'].str.replace('.0$','',regex=True)
```

```python
gpd_alicante
```

```python
gdf_alicante_habitantes = pd.merge(gpd_alicante,df_habitantes,how='inner',left_on='CODIGOINE',right_on='INE')
```

```python
gpd_municipios.to_crs(pyproj.CRS.from_epsg(4326), inplace=True)
```

```python
fig = px.choropleth(gdf_alicante_habitantes,
                    geojson=gdf_alicante_habitantes.geometry,
                    locations=gdf_alicante_habitantes.index,
                    color='2021',
                    hover_name='Municipio',
                    color_continuous_scale="turbid",
                    labels={'2021':'Habitantes'}
                    )
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.show()
```
